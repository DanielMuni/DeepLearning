```python
import pandas as pd
from sklearn.decomposition import PCA
from sklearn.preprocessing import OneHotEncoder, LabelEncoder
from dataprep.eda import create_report
```

```python
df = pd.read_csv('../../data/telecom_churn_me.csv')
df
```

```python
df_a = df.drop(columns= ['PTY_PROFILE_SUB_TYPE','SOCIO_ECONOMIC_SEGMENT','PARTY_NATIONALITY','PARTY_GENDER_CD','STATUS'])
df_b = df[['PTY_PROFILE_SUB_TYPE','SOCIO_ECONOMIC_SEGMENT','PARTY_NATIONALITY','PARTY_GENDER_CD','STATUS']]

df_a
```

```python
df_b = df_b.apply(LabelEncoder().fit_transform)
df_b
```

```python
for i in ['PTY_PROFILE_SUB_TYPE','SOCIO_ECONOMIC_SEGMENT','PARTY_NATIONALITY','PARTY_GENDER_CD','STATUS']:
    df_a[i] = df_b[i]
df_a
```

```python
df_pca = df_a.drop(columns = ['Unnamed: 0'])
df_pca
```

```python
pca_data = PCA(n_components = 27).fit_transform(df_pca)
pca_data
```

```python
names = ["TARGET","YEAR_JOINED","CURRENT_YEAR","BILL_AMOUNT","PAID_AMOUNT","PAYMENT_TRANSACTIONS","PARTY_REV","PREPAID_LINES","POSTPAID_LINES","OTHER_LINES","LINE_REV","MOUS_TO_LOCAL_MOBILES","MOUS_FROM_LOCAL_MOBILES","MOUS_TO_LOCAL_LANDLINES","MOUS_FROM_LOCAL_LANDLINES","MOUS_TO_INT_NUMBER","MOUS_FROM_INT_NUMBER","DATA_IN_BNDL","DATA_OUT_BNDL","DATA_USG_PAYG","COMPLAINTS","Years_stayed","PTY_PROFILE_SUB_TYPE","SOCIO_ECONOMIC_SEGMENT","PARTY_NATIONALITY","PARTY_GENDER_CD","STATUS"]
imputed_pca_df = pd.DataFrame(pca_data, columns = names)
imputed_pca_df
```

```python
report = create_report(df_a, title='Telco Customer Churn data report after labelenconder')
report.show_browser()
```
